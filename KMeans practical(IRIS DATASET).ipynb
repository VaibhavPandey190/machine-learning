# Elbow technique to find number of targets or clusters or KMeans

```python

# K = 1 to 10, we calculate distortion, and then we plot the graph between K and distortions

distortions = []#here distortion will be calculated for every value of k that is here 1 to 9

K = range(1,10)

```

```python
from scipy.spatial.distance import cdist
import math
```

```python
iris_df.head()
X = iris_df.loc[:,['petal_length','petal_width']]#any two can be taken
X = np.array(X)
X
```

```python
for k in K:
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeans_model.cluster_centers_, 'euclidean'), axis=0) / X.shape[0]))
#     axis can be 0 and 1 both
```

```python
distortions#distortions from 1 to 9 value of k
```

```python
#plot distortion againts K value 
plt.plot(K, distortions, 'green')#color canbe green,red etc or g+that is green colour with +as linestyle etc
plt.xlabel("K - Number of Clusters")
plt.ylabel("Distortions")

```
